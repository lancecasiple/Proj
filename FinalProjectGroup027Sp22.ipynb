## Background & Prior Work

In our globalized world, American foreign aid has taken on immense importance as a way to promote economic stability, growth, build relationships, as well as project influence on nations throughout the world. [^Michel] President Biden’s budget for the 2023 fiscal year requests more than $17 billion for the US Department of State and its foreign aid development agency, USAID, and $11 billion for the US Department of Health and Human Services. [^whitehouse.gov] With these being central goals of foreign aid, measuring their actual impact in recipient countries is important to understand its effectiveness. Average life expectancy is a useful measure of the quality of life for a sample population. It is defined as a measure “based on an estimate of the average age that members of a particular population group will be when they die.” [^Ortiz-Ospina]

By comparing the impact of foreign aid recieved by the US to the country's average life expectency, we hope to gain a clearer picture on the effectiveness of foreign aid.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
oecd_total_reciepts = pd.read_csv('data/OECD Total Reciepts.csv.gz', compression='gzip')
oecd_total_reciepts.head()
```

# Data Cleaning

Describe your data cleaning steps here.

```python
# Cleanup us_foreign_assistance by removing uneeded columns
us_foreign_assistance.drop(['Country ID',
                            'Region ID',
                            'Income Group ID',
                            'Managing Sub-agency or Bureau ID', 
                            'Managing Sub-agency or Bureau Acronym',
                            'Managing Sub-agency or Bureau Name',
                            'Operating Unit',
                            'US Category ID',
                            'US Category Name',
                            'US Sector ID',
                            'US Sector Name',
                            'Funding Acount ID',
                            'Funding Agency Name',
                            'Funding Agency ID',
                            'OCO Flag',
                            'Activity ID',
                            'Activity Description',
                            'Transaction Type ID'], axis=1, inplace=True)

# Drop all rows related to 'World' funding
us_foreign_assistance = us_foreign_assistance[us_foreign_assistance['Country Name'] != 'World']
```

```python
# Cleanup un_period_indicators by removing undeeded columns
un_period_indicators.drop(['Births',
                           'Deaths',
                           'DeathsMale',
                           'DeathsFemale',
                           'CNMR',
                           'NetMigrations',
                           'VarID'], axis=1, inplace=True)
```

```python
WHO_life_expectancy.columns = [i.replace('-','_') for i in WHO_life_expectancy.columns]
WHO_life_expectancy.drop(['percentage expenditure',
                            'under_five deaths ',
                            'Total expenditure',
                            ' thinness  1_19 years',
                            'Income composition of resources',
                            'Schooling',' thinness 5_9 years'], axis=1, inplace=True)
#WHO_life_expectancy.columns
```

```python
un_period_indicators_avgs = un_period_indicators.groupby(['Location', 'Time']).mean().reset_index()
un_period_indicators_avgs['MidPeriod'] = un_period_indicators_avgs['MidPeriod'].astype(float)
un_period_indicators_avgs
```

```python
# Group the US Data by Country, Fiscal Year, and Category
country_metadata = us_foreign_assistance.groupby(['Country Name', 'Fiscal Year', 'International Category Name']).agg({'current_amount':'sum'})
country_metadata.columns = ['Total Funds']
country_metadata = country_metadata.reset_index()
```

```python
WHO_2015 = WHO_life_expectancy[WHO_life_expectancy.Year == 2015]
WHO_2015 = WHO_2015.rename(columns={"Life expectancy ":'Life_expectancy'})
UFA_2015 = us_foreign_assistance[us_foreign_assistance['Fiscal Year'] == 2015]
UFA_2015 = UFA_2015.rename(columns={'Country Name':'Country'})
UFA_2015 = UFA_2015.groupby(['Country']).sum()
merged_WHO_UFA = UFA_2015.merge(WHO_2015, on='Country')
merged_WHO_UFA = merged_WHO_UFA[(merged_WHO_UFA.current_amount < 4000000000) & (merged_WHO_UFA.current_amount > 0)]
```

# Data Analysis & Results

First, let's find the most funded country by US foreign assistance using the dataframes creatd from above. The most country who recieves the most funds from the US should offer a clear insite and answer to our hypothesis.

```python
# Group the US Data by Country, and Fiscal Year
country_metadata_year = us_foreign_assistance.groupby(['Country Name', 'Fiscal Year']).agg({'current_amount':'sum'})
country_metadata_year.columns = ['Total Funds']
country_metadata_year = country_metadata_year.reset_index()

# Output the most funded county
country_metadata_year.loc[country_metadata_year['Fiscal Year'].idxmax()]
```

Afganistan recieved a large quantity of aid this year, especially due to recent conflicts. Lets double check that is the most funded by printing the top ten highest funded countries.

```python
country_metadata_year.sort_values(by=['Total Funds'], ascending=False).head(10)
```

It's clear Afghanistan has been a longterm investment in foreign aid from the US. This makes them a perfect candidate to test our hypothesis on the impact of US Foreign Aid having a positive impact on a recipient country's life expectency.

```python
# Plot the change in Total Funds by International Category per Fiscal Year for Afghanistan
plt.figure(figsize=(15, 8))
sns.set(style='ticks')
ax = sns.lineplot(data=country_metadata_year[country_metadata_year['Country Name'] == 'Afghanistan'], 
                  x='Fiscal Year',
                  y='Total Funds')

# Plot the Avg Life Expectency by time period
ax2 = ax.twinx()
sns.lineplot(data=un_period_indicators_avgs.loc[(un_period_indicators_avgs['Location'] == 'Afghanistan') & (un_period_indicators_avgs['MidPeriod'] > 2000) & (un_period_indicators_avgs['MidPeriod'] < 2025)],
            x='MidPeriod',
            y='LEx',
            linestyle='dashed',
            marker='o', 
            markersize=10,
            lw=3,
            ax=ax2,
            label='Avg. Life Expectency')
ax2.legend(bbox_to_anchor=(1.02, 0.65), loc='upper left', borderaxespad=0)
ax.set_xlim(2006, 2020)
ax2.set_ylabel('Life Expectency')

ax.set(title='Comparing Change in US Aid per Year to Avg. Life Expectency of for Afghanistan')
plt.show()
```

This graphs display a solid line of the total funds per year given to Afghanistan compared to Afgahnistan's life expectency displayed as a dashed line. While this does give some interesting insight into Afgahnistan's foreign aid compared to the country's average life expectency, it also brings up a few questions. 

Most importantly, how is this foreign aid broken down? Just because a lot of aid was given (for example look at 2011), which saw a large spike in aid. Not forms of aid should be expected to have as direct of an impact on life expectency as other.

It is also worth noting that it would be unrealistic to expect a 1:1, or immediate and direct impact of aid on life expectency during the same year.

```python
# Plot the change in Total Funds by International Category per Fiscal Year for Afghanistan
plt.figure(figsize=(15, 8))
sns.set(style='ticks')
ax = sns.lineplot(data=country_metadata[country_metadata['Country Name'] == 'Afghanistan'], 
                  x='Fiscal Year',
                  y='Total Funds',
                  hue='International Category Name')
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)

# Plot the Avg Life Expectency by time period
ax2 = ax.twinx()
sns.lineplot(data=un_period_indicators_avgs.loc[(un_period_indicators_avgs['Location'] == 'Afghanistan') & (un_period_indicators_avgs['MidPeriod'] > 2000) & (un_period_indicators_avgs['MidPeriod'] < 2025)],
            x='MidPeriod',
            y='LEx',
            linestyle='dashed',
            marker='o', 
            markersize=10,
            lw=3,
            ax=ax2,
            label='Avg. Life Expectency')
ax2.legend(bbox_to_anchor=(1.02, 0.65), loc='upper left', borderaxespad=0)
ax.set_xlim(2006, 2020)
ax2.set_ylabel('Life Expectency')

ax.set(title='Comparing Change in US Aid by Category per Year for Afghanistan to Avg. Life Expectency')
plt.show()
```

In this graph, we try to explore the relationship between the amount of aid the U.S. send to Afghanistan in terms of (Argriculture, Governance, etc) and the life expectancy of Afghanistan. We separated each categories of aid to have a better understanding of which type of aid might have a more significant impact on the life expectancy than others. In this graph, we can see that the U.S. send a significant amount of governance money support and infrastructure support. We also see a general upward trends for life expectancy despite the amount of aid fluctuates yearly. Especially in the year 2011 - 2012, we see a big dip in the amount of governance aid and the life expectancy still increases. From this graph, we suspect that foreign aid might have some impact on Afghanistan life expectancy but we are still not 100% sure and we would need to explore the data further.
